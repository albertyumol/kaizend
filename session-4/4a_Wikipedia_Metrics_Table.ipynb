# Wikipedia Metrics Table
# 4a_Wikipedia_Metrics_Table.ipynb
### Make CSV file with the ff. Wikipedia activity metrics of Forbes 100 celebrities pages:
* edits = no. of edits
* size = total size of edits
* users = no. of contributing users

# Import libraries

```python
import pandas as pd
import numpy as np
```

# Define output file path & filename

```python
fn_out = "/root/Downloads/DIY_Data_Analytics-20200116T140155Z-001/DIY_Data_Analytics/output/wikipedia_metrics.csv"
```

# Read in data from CSV file

```python
fn = "/root/Downloads/DIY_Data_Analytics-20200116T140155Z-001/DIY_Data_Analytics/input/wikipedia_edits.csv.zip"
df = pd.read_csv(fn)
```

Count no. of rows

```python
len(df)
```

Inspect first rows

```python
df.head()
```

# Aggregate Using Groupby

1. Get the total no. of Wikipedia edits per celebrity

```python
df_edits = pd.DataFrame(df.groupby("title").size().reset_index(name='edits')) #["revid"].agg([len])
```

```python
df_edits.head()
```

Rename column to edits

```python
#df_edits.columns = ["edits"]
df_edits.head()
```

```python
df_edits.columns = ["".join(v) for v in df_edits.columns.values]
```

```python
df_edits.columns# = ['%s%s' % (a, '|%s' % b if b else '') for a, b in df_edits.columns]
```

Inspect top celebrities by no. of edits - this replicates what we got earlier from df["title"].value_counts()

```python
df_edits.sort_values(by="edits", ascending=False).head(5)
```

2. Get the total size of Wikipedia edits per celebrity

```python
df_size = pd.DataFrame(df.groupby("title")["size"].sum())
df_size.columns = ["size"]
df_size.head()
```

```python
df.groupby("title")["size"].agg([len, np.sum, np.mean, np.median, np.std]).head(40)
```

```python
df.groupby("title")["size"].agg([len, np.sum, np.mean, np.median, np.std]).head(82).tail(42)
```

Inspect top celebrities by total size of edits

```python
df_size.sort_values(by="size", ascending=False).head(5)
```

3. Get no. of (unique) Wikipedia users who contributed edits per celebrity

```python
df_users = df.groupby(["title"])["user"].nunique()
df_users = pd.DataFrame(df_users)
df_users.columns = ["users"]
```

```python
df_users.head()
```

Inspect top celebrities by no. of contributing users

```python
df_users.sort_values(by="users", ascending=False).head(5)
```

# Merge results into 1 dataframe

First, reset index of each dataframe, then merge on title column

```python
df_edits.reset_index(inplace=True)
df_size.reset_index(inplace=True)
df_users.reset_index(inplace=True)
```

```python
df_edits.head()
```

```python
df_out = pd.merge(df_edits, df_size)
df_out = pd.merge(df_out, df_users)
df_out.head()
```

# Save dataframe to CSV file

```python
df_out.to_csv(fn_out, index=False)
```
